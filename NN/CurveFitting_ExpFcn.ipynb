```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit


# Experimental data
x = np.array([0, 0.5, 1, 2, 3, 4, 5, 6, 8, 10, 12, 14, 16, 18])
y = np.array([0.2380, 0.2266, 0.2847, 0.3717, 0.4336, 0.4201, 0.4380, 0.6287, 0.6946, 0.7523, 0.7966, 0.8962, 0.8488, 1])

# Define the function for the trendline
def trendline(x, a, b):
    return a*np.exp(-3*b*12 / (b*12 + 12800*x))

# Fit the data to the trendline function
popt, _ = curve_fit(trendline, x, y)

# Extract the optimized values of a and b
a_opt, b_opt = popt

# Generate y values for the trendline
trendline_y = trendline(x, a_opt, b_opt)

# Equation for the logarithmic trendline
equation = f'y = {a_opt:.4f}*exp(- 3* {b_opt:.4f}*12 / ({b_opt:.4f}*12 + 12800*x))'

# Plot the data and the trendline
plt.scatter(x, y, label='Data')
plt.plot(x, trendline_y, color='red', label='Trendline')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Trendline Fit')
plt.text(2, 0.2, equation, fontsize=12, color='blue')
plt.legend()
plt.grid(True)
plt.show()
```
